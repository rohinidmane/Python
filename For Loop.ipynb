```python
#For Loop----(iteration--- repeatition)
sample=[1,2,3,"ram"]
for i in sample:
  print(i)
```

```python
list1 = []  # empty list
for i in "Python"[::-1]:
  list1.append(i)
  print(list1)
```

```python
list1=[]
for i in "Python"[::-4]:
  list1.append(i)
print(list1)
```

```python
for i in range(0,26):
  print(i**3)
```

```python
country_city_rever_list=[('India','New Delhi','Gang'),('Austrelia','canberra','Rovers'),('USA','Washington','Missouri')]
for country,city,river in country_city_rever_list:
  print(f'The capital of the country {country} is new {city} and also has the river {river}')
```

Enumerate

```python
name=['ram','Neelam','sham']
for i in enumerate(name,start=1):
  print(i)
```

```python
list(enumerate(name))
```

```python
#use of zip
names = ['asha', 'abhi', 'ram']
age = [20, 89, 23]
zip(names, age)
```

```python
info = list(zip(names, age))
print(info)
```

```python
#Nested for loop
legends=['rajni','kamal','vikram']
movies=['movie1','movue2','movie3']
for x in legends: # outer for loop
  for y in movies: # inner for loop
    print(x,y)
```

```python
for i in range(1,4):
  for y in range(1,4):
    print(f'{i} x {y}={i*y}')
  print('-------------')
```
