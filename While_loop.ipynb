While Loop

```python
i = 1
while i<6:
  print(i)
i = i + 1 # i+= 1
```

```python
total = 0
num = int(input("enter a number(0 to stop:)"))
while num != 0:
  total += num # total = total + num
  num = int(input("enter a number to (0 to stop: "))
print("Total: ", total)
```

Nested While Loop

```python
i = 1
while i <= 3: # outer while loop
  j = 1
  while j <= 3: # inner while loop
    print(f"i={i}, j={j}")
    j += 1
  i += 1
```
