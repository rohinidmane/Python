```python
l = [1,2,3,3,4,5,56,88,9,8,5,2]

def if_even(l):
    e = []
    for i in l:
        if i % 2 == 0:
            e.append(i)
    return e
```

**Yield**

```python
#yield using for loop
def even_generator(limit):
    for i in range(2, limit+1, 2):
        return i

print(even_generator(10))
```

```python
def even_generator(limit):
    for i in range(2, limit + 1, 2):
        yield i

#print(even_generator(10))
for num in even_generator(10):
    print(num)
```

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for number in count_up_to(3):
    print(number)
```

**Recursive** **Function**

```python
def factorial(n):
    if n == 0:
        return 1  # Base case
    else:
        return n * factorial(n-1)  # Recursive call

# get the user input
num = int(input("Enter a non negative integer "))
result = factorial(num)
print(f'The Factorial of {num} is', result)
```

```python
factorial ((6)) # Here, factorial(5) calls factorial(4), then factorial(3), and so on, until it reaches factorial(0).
```

```python
# Sum of first n natural numbers
def sum_natural(n):
    if n == 1:
        return 1  # Base case
    else:
        return n + sum_natural(n-1)  # Recursive call sum_natural(5) → 5 + 4 + 3 + 2 + 1 = 15
```

```python
sum_natural(9)
```
