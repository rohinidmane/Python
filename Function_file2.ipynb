**Variable Length Arguments**

```python
# *args --- collect all positional arguments as a tuple
def print_marks(*args):  #marks  *---catch all
  print("Marks: ", args)
  return sum(args)

print_marks(78, 89, 90, 45,90,78,10,90,100,78)
```

```python
def myfun(*argv):
  for arg in argv:
    print(arg)

myfun('Hello', 'Welcome', 'To', 'Cloudblitz')
```

```python
# **kwargs=== collects all named arguments as a dictionary
def print_student_info(**kwargs):
  print("Student Information: ", kwargs)

print_student_info(name = 'Anu', age = 23, course = 'Math',college ="SRIT")
```

**Anonymous Function --- Lambda**

```python
# lambda
square = lambda x: x * x # x = 5
print(square(5))
```

```python
add = lambda a, b, c: a + b + c
print(add(5,8,10))
```

**Higher Order Function**

```python
# higher order function
def greet(name):
  return f"Hello, {name}!"

def process_name(func, name):
  return func(name)

# calling higher-order function
print(process_name(greet, "Alice"))
```

**Map ---- map(func, iterable)**

```python
numbers = [1,2,3,4]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

```python
a = [1,2,3]
b = [4,5,6]
c = [9,7,6]
result = map(lambda x, y,z: x + y+ z, a, b,c)
print(list(result))
```

**Filter ()--- Basic Syntax --- filter(function, iterable)**

```python
numbers = [1,2,3,4,5,6,7]

even_numbers = filter(lambda x: x%2 == 0, numbers)
print(list(even_numbers))
```

```python
def is_positive(x):
  return x > 0

data = [-3, 9, -2, 0, 2, 4]
result = filter(is_positive, data)
print(list(result))
```

**Sorted ()**

```python
# sorting words based on length

a = ["Python", "java", "javascript"]

res = sorted(a, key = len)
print(res)
```

```python
students = [("Alice", 85), ("Gob", 78), ("Charlie", 90)]
sorted_students = sorted(students, key = lambda x: x[1],reverse=True)
print(sorted_students)
```

**reduce()**

```python
from functools import reduce
numbers = [1,2,3,4,5,6]

result = reduce(lambda x, y: x + y, numbers)
print(result)
```

```python
from functools import reduce
numbers = [1,2,3,4,5,6]

multiply = reduce(lambda x, y: x * y, numbers)
print(multiply)
```

```python
# Function to count vowels in a string
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

print(count_vowels("Hello World"))
```

```python
# Function to calculate area of a circle
def area_of_circle(radius):
    pi = 3.14
    area = pi * radius * radius
    return area
print(area_of_circle(7))
```

```python
# Function to perform basic calculator operations
def calculator(a, b, operation):
    if operation == 'add':
        return a + b
    elif operation == 'subtract':
        return a - b
    elif operation == 'multiply':
        return a * b
    elif operation == 'divide':
        if b != 0:
            return a / b
        else:
            return "Error: Division by zero"
    else:
        return "Invalid operation"





print(calculator(10, 5, 'add'))
print(calculator(10, 5, 'divide'))
```
