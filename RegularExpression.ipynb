**Importing regular expression re module**

```python
import re
```

**re.search()**

```python
import re
regex = r"([a-zA-Z]+) (\d+)" # r is raw string literal

match = re.search(regex, "I was born on June 2024")
if match != None:
  print("Match at the index %s, %s" % (match.start(), match.end()))
  print("Full match: %s" % (match.group(0)))  # june 2024
  print("Month: %s" % (match.group(1)))
  print("Day: %s" % (match.group(2)))

else:
  print("The regex pattern does not match")
```

```python
text = "The rain in Spain falls mainly in the plain"

# search for the pattern
pattern = r"Spain"
match = re.search(pattern, text)
if match:
  print("Match found:", match.group())

else:
  print("No Match")
```

**re.findall()**

```python
text = "The rain in Spain falls mainly in the plain"
# find all occurances of pattern
#pattern= r"in"
pattern = r"\bin\b"
matches = re.findall(pattern, text)
print("All Matches:", matches)
```

```python
text = '''Cloudblitz manager phone number is 9873746466, call him if you have any questions regarding sessions management.
Office landline number is (879)-773-83773'''

pattern = r'\d{10}|\(\d{3}\)-\d{3}-\d{5}'

matches = re.findall(pattern, text)
matches
```

```python
import re
text = "ababab"
pattern = re.findall(r"aba", text)
print(pattern)

pattern = re.findall(r"(?=(aba))", "ababab")
print(pattern)
```

**re.split()**

```python
from re import split

print(split(r'\W+', "Words, words, Words"))
print(split(r'\W+', "Word's, words, Words"))
print(split(r'\W+', 'On 12th Jan 2016, at 11:02 AM'))

print(split(r'\d+','On 12th Jan 2016, at 11:02 AM'))
print(split(r'\d+','On 12th Jan 2016, at 11:02 AM', 2)) # here 2 is maxsplit
```

**re.sub()**

```python
import re
print(re.sub('ub', '~*', 'Subject has Uber booked already',flags = re.IGNORECASE))
print(re.sub('ub', '~*','Subject has Uber booked already'))
print(re.sub('ub', '~*','Subject has Uber booked already', count = 1, flags = re.IGNORECASE))
```

**re.match()**

```python
import re
result = re.match(r'\d+', 'abc111')
if result:
  print("Matched:",result.group())
else:
  print("Not found")
```
