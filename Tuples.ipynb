Slicing of tuple

```python
tup = (1,2,3,4)
print(tup[::-1]) # reverse
```

```python
t = (9,3,2,5,7,3,7)
print(t[2:7])
```

```python
t = (1, 2, 3, 4, 5, 6)
print(t[::2])
```

```python
t = (10, 20, 30, 40, 50)
print(t[-4:-1])
```
