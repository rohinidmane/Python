Slicing:Accesing the element using slicing

```python
age = [20, 78, 90, 100, 25, 76]
age

print(age[1:5])

print(age[-5: -2])
```

```python
numbers = [1, 2, 3, 4, 9, 7, 8] # [start(inclusive), stop(exclusive), step]

print(numbers[:6])
print(numbers[2: 5])
print(numbers[-3: -1])
print(numbers[-1:-3])
```

Adding elements to the python list

```python
#Append()
name =["raj","sumit","rohini"]
name.append("Vidya")
print(name)
```

```python
#insert()
fruits =["orange","mango","banana"]
print(fruits)
fruits.insert(2,"Guava")
fruits.insert(1,['lemon','watermelon'])
print(fruits)
```

```python
#extend
fruits =["orange","mango","banana"]
fruits.extend(["Guava","watermelon","strawberry"])
print(fruits)
```

```python
# .append() and .extend()
mixed_list = [1,2,3,4,'data', [4,2,['sita','mahima']],'😎','🎂',3.14]
mixed_list.append(['logic', 'gate'])
print(mixed_list)
```

```python
mixed_list[5][2][1]
```

```python
mixed_list.extend(['data', 'science'])
print(mixed_list)
```
